# Text Preprocessing

**1. Saring kata yang bahasa inggris saja**

```python
def is_english(text):
    if not text:
        return False
    lang, _ = langid.classify(text)
    return lang == 'en'
```

```python
english_reviews = data_review[data_review['review'].apply(is_english)]
```

```python
for review in english_reviews["review"]:
    print(review)
```

Dari hasil diatas memang tidak 100% data yang diproses outputnya adalah bahasa inggris, terdapat beberapa kalimat full bahasa indonesia, namun hal itu jauh lebih baikk daari sebelumnya. Lalu kata inggris seperti game, contoh pada kalimat "coba game online" tidak dihapus karena kata game adalah bahasa inggris

**2. Bersihkan teks dari angka, simbol, link, emoji, dan sebagainya lalu buat huruf semuanya jadi kecil**

```python
# Fungsi untuk membersihkan text dari angka, simbol, link, dan lain-lain
def cleaningtext(review):
  review = re.sub(r"@[A-Za-z0-9]+", "", review)
  review = re.sub(r"#[A-Za-z0-9]+", "", review)
  review = re. sub(r"RT[\s]", "", review)
  review = re.sub(r"http\S+", "", review)
  review = re.sub(r"[0-9]+", "", review)
  review = re.sub(r"[^\w\s]", "", review)
  review = review.replace("\n", " ")
  review = review.translate(str.maketrans("", "", string.punctuation))
  review = review.strip(" ")
  return review

# Fungsi untuk mengbuah huruf menjadi keci
def casefolding(review):
  review = review.lower()
  return review

# Apply
english_reviews["review_clean"] = english_reviews["review"].apply(cleaningtext)
english_reviews["review_casefolding"] = english_reviews["review_clean"].apply(casefolding)
```

**3. Tokenisasi**

```python
# membuat fungsi tokenisasi
def tokenizingtext(review):
  review = word_tokenize(review)
  return review
# apply
english_reviews["review_token"] = english_reviews["review_casefolding"].apply(tokenizingtext)
```

**4. Menghapus kalimat bahasa indonesia dan kalimat dengan mix bahasa**

```python
# kamus
english_vocab = set(w.lower() for w in words.words())
# Membuat fungsi untuk cek token berbahasa inggris
def english(tokens):
    if not tokens or not isinstance(tokens, list):
        return False
    tokens = [token for token in tokens if token.isalpha() and token.strip() != '']

    if not tokens:
        return False
    return all(token in english_vocab for token in tokens)
```

```python
english_reviews2 = english_reviews[english_reviews['review_token'].apply(english)]
```

```python
for review in english_reviews2["review_token"]:
  print(review)
```

Jika dilihat dari hasil diatas jauh lebih baik dari proses penyaringan bahasa dimana bahasa inggris saja yang digunkan. Hasil diatas tidak terdapat kata berbahasa indonesia, thai, atau lainnya kecuali inggris

**5. Stopwords**

```python
# Membuat Fungsi Stopwords
stop_words = set(stopwords.words('english'))
# Fungsinya
def remove_stopwords(tokens):
  return [token for token in tokens if token.lower() not in stop_words]
# apply
english_reviews2["review_stopwords"] = english_reviews2["review_token"].apply(remove_stopwords)
```

```python
english_reviews2["review_stopwords"]
```

```python
# Membuat fungsi menjadi kalimat
def toSentence(list_words):
  sentence = " ".join (word for word in list_words)
  return sentence
# apply
english_reviews2["review_final"]=english_reviews2["review_stopwords"].apply(toSentence)
```

# Pelabelan

```python
# Membuat Dataframe baru
data_sentiment = english_reviews2[["review_final"]].copy()
data_sentiment.shape
```

```python
analyzer = SentimentIntensityAnalyzer()
def sentiment_label(text):
    score = analyzer.polarity_scores(text)["compound"]
    if score >= 0.05:
        return "positive"
    elif score <= -0.05:
        return "negative"
    else:
        return "neutral"
```

Kode diatas menggnkan Vader untuk proses pelabelan data dimana fungsi sentiment_label() digunakan untuk mengklasifikasikan teks ke dalam kategori sentimen.
- Jika skor lebih dari atau sama dengan 0.05, dianggap positif

- Jika skor kurang dari atau sama dengan -0.05, dianggap negatif

- Sisanya dianggap netral

```python
data_sentiment["sentiment_label"] = data_sentiment["review_final"].apply(sentiment_label)
```

# Data Splitting, Ekstraksi Fitur dan Modelling

- Encoding pada kolom Label untuk proses modeling

```python
le = LabelEncoder()
data_sentiment["sentiment_label"] = le.fit_transform(data_sentiment["sentiment_label"])
for i, label in enumerate(le.classes_):
    print(f"{i} -> {label}")
```

Modelling menggunakan MLP (Multilayer Perceptron) dengan spliting (80:20) dan TF-IDF**

```python
# Bagi data menjadi untuk train dan test
x_1 = data_sentiment["review_final"]
y_1 = data_sentiment["sentiment_label"]
# ekstraksi feature
tfidf = TfidfVectorizer(max_features=1000, min_df=10, max_df=0.8 )
x_tfidf = tfidf.fit_transform(x_1)

x_tfidf_dense = x_tfidf.toarray()
```

```python
# perbandingan data
x_train, x_test, y_train, y_test = train_test_split(x_tfidf_dense, y_1, test_size=0.2, random_state=42)
```

```python
model = Sequential([
    Dense(512, activation='relu', input_dim=x_tfidf.shape[1]),
    Dropout(0.5),
    Dense(256, activation='relu'),
    Dense(128, activation='relu'),
    Dense(3, activation='softmax')
])

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
```

```python
# One hot encoding pada label untuk proses melatih model
y_train_encoding1 = to_categorical(y_train, num_classes=3)
y_test_encoding1 = to_categorical(y_test, num_classes=3)
```

```python
# Meleatih model
latih_model = model.fit(
    x_train, y_train_encoding1,
    epochs=20,
    batch_size=64,
    validation_data=(x_test, y_test_encoding1)
)
```

```python
loss, accuracy = model.evaluate(x_test, y_test_encoding1)
print(f"Loss : {loss:.4f}")
print(f"Akurasi : {accuracy:.4%}")
```

Model mempunyai akurasi yang baik yaitu hampir 94%. jadi bisa dibilang model cukup pintar dalam memahami sentimen dari data yang diuji.
