```python
# Import library goole_play_scrapper untuk mengakses ulasan dan informasi
from google_play_scraper import app, reviews, Sort, reviews_all
import pandas as pd
import csv
```

```python
# Mengambil ulasan dari aplikasi di playstore
scrapreview = reviews_all(
    'com.linecorp.LGGRTHN',
    lang = 'en',
    sort = Sort.MOST_RELEVANT,
    count=1000
)
```

```python
with open('raw_data_lgr.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['review'])
    for review in scrapreview:
        writer.writerow([review['content']])
```

```python
app_reviews_df = pd.DataFrame(scrapreview)
app_reviews_df.shape
app_reviews_df.head()
app_reviews_df.to_csv('raw_data_lgr.csv', index=False)
```
